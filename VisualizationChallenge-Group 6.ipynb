```python
import pandas as pd
import numpy as np

df=pd.read_excel('GainesvilleHomes.xlsx')
df
```

Here we created a scatterplot comparing the square footage of a home to the price with the line of best fit. Pros: We can now see that there is a positive correlation between the size of a home and the prices. The con is that there are other factors that can effect the price of a home but they are not shown here, this graph shows a 1:1 relationship between two variables.

```python
sns.lmplot(x='SQUARE_FOOTAGE', y='Price(thousands)', data=df, fit_reg=True)   
```

Here we created a Point Plot comparing the number of bed and bathrooms to the Price of the home. Pros: This graph makes it easy to see the coorelation between the two varibale but it also allows home-seekers to see that there are price variations with the number of bedroom & bathrooms.
A cons is that there are other variables that should be taken into consideration here and could effect the price more.

```python
sns.pointplot(x='BEDS_BATHS', y='Price(thousands)', data=df)
```

This graph shows how different factors relate to others. Pro: You can easily see the relationship between different variables and choose the one that fits a specific criteria. Con: There is a lot of information here and it may be difficult to locate you data of interest. 


```python
sns.pairplot(df, height=2.5)
```

This graph shows the relationship between lot size and price. 
Pro: Again, this graph looks at a relationship between two variables of interest.
Con: Although this graph is simple to look at, it is not the only factor which contributes to the price of the home. Other factors are also important to consider too.

```python
sns.lmplot(x='LOT SIZE', y='Price(thousands)', data=df, fit_reg=True)   
```

The heatmap was not chosen because it didn't really apply here. All of the variables had different effects on the price. Price was the most important variable in this scenario because, ultimately home-seekers consider the price to make sure they are in budget.
