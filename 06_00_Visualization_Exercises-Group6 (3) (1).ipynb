```python
import matplotlib.pyplot as plt
import pandas as pd
gspc = pd.read_csv("GSPC.csv")
gspc['Date'] = pd.to_datetime(gspc['Date'])
gspc.head()
```

```python
gspc.set_index('Date', inplace = True) 
gspc
```

```python
gspc['High'].resample(rule = 'W').median().plot(kind = 'line', figsize = (14,6), marker = 'o')
gspc['Low'].resample(rule = 'W').median().plot(kind = 'line', figsize = (14,6), marker = '+')
plt.title('Weekly Median of High and Low price for S&P')

plt.xlabel('Years', fontsize=12)
plt.ylabel('USD', fontsize=12)
plt.legend()
```

In the chart above, we have found the mean of the Low and High prices for the S&P

```python
import matplotlib.pyplot as plt
import pandas as pd
surv = pd.read_csv("covid_survey_full.csv")
surv.head()
```

```python
surv.info()
```

```python
surv_income = surv['income']

plt.style.use('ggplot')

binsep = [0, 20, 40, 60, 80, 100, 120]

plt.hist(surv_income, bins=binsep, edgecolor='black')
plt.title("Income Range of Survey Takers")

plt.xlabel('Income (In Thousands)', fontsize=11)
plt.ylabel('Number of Survey Takers', fontsize=11)
```

In the Histogram above, we are looking at the number of survey takers with regards to their income range. Here we changed the bin amount on the x axis, edited the edgecolor, changed the style of the histogram from the default and changed the font size of the x and y labels to better fit the graph.

```python
import matplotlib
from matplotlib import transforms, pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
seg = pd.read_csv("segments.csv")
seg.head()
```

The graph shows the relationship between age and income

```python
# Use scatterplot() to generate a scatter plot
plt.title("sns.scatterplot()")
sns.scatterplot(x="Income", y="age", hue="Segment", data=seg)
plt.title("Income VS Age")

# Use relplot() to generate a scatter plot
g1 = sns.relplot(x="Income", y="age", hue="Segment", data=seg, kind='scatter')
g1.fig.suptitle("sns.relplot()");
plt.title("Income VS Age")
```

```python
# change markers
plt.figure(figsize=(160,160))
sns.relplot(x="Income", y="age", hue="HomeOwnership", style="HomeOwnership",
            data=seg);
plt.title("Income VS Age in relationship to Home Ownership")
```

```python
import matplotlib
from matplotlib import transforms, pyplot as plt
import numpy as np
import seaborn as sns
seg = pd.read_csv("segments.csv")
seg.head()
```

```python
df = seg['Segment'].value_counts()
df
```

```python
df = seg.groupby('Segment').size()
df
```

```python
slices = df
#plt.pie(slices)
labels = ['ExSuburban', 'ExUrban', "Suburban", "Urban"]

plt.pie(slices, labels=labels)
plt.title("Proportion of houses owned in different segments")
plt.show()
```

The chart shows the proportion of houses owned in different segments: Urban, ExUrban, ExSuburban, and Suburban.

```python
df2 = seg.groupby('Segment').size().to_frame(name='count').reset_index()
df2
```

```python
df3=df2.set_index(['Segment'], append = False)
df3
```
